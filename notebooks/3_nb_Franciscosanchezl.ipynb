```python
def cambio_mes (param):
    if param == "Ene":
        param = "Jan"
    if param == "Feb":
        param = "Feb"
    if param == "Mar":
        param = "Mar"
    if param == "Abr":
        param = "Apr"
    if param == "May":
        param = "May"
    if param == "Jun":
        param = "Jun"
    if param == "Jul":
        param = "Jul"
    if param == "Ago":
        param = "Aug"
    if param == "Sep":
        param = "Sep"
    if param == "Oct":
        param = "Oct"
    if param == "Nov":
        param = "Nov"
    if param == "Dic":
        param = "Dec"
        
    return param
```
